```python
# Calculating the mean vote
meanvote = dataset['vote_average'].mean()
print(meanvote)
```

```python
#Calculating the cutoff value (minimum vote)
min_vote = dataset['vote_count'].quantile(0.90)
print(min_vote)
```

```python
#Filter the movies into other Dataframe
best_movies = dataset.copy().loc[dataset['vote_count'] >= min_vote]
best_movies.shape
```

```python
#Computing the weighthed average
def weighted_rating(x, min_vote=min_vote, meanvote=meanvote):
    voters = x['vote_count']
    avg_vote = x['vote_average']
    return (voters/(voters+min_vote) * avg_vote) + (min_vote/(min_vote+voters) * meanvote)
```

```python
best_movies['score'] = best_movies.apply(weighted_rating, axis = 1)
```

```python
best_movies = best_movies.sort_values('score', ascending = False)
```
